```python
#basic mathematical library for plotting graphs and analysing data
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60]
y = [23,11,9,37,12,10]
```

```python
#pass two arrays as input arguments to plot()
plt.plot(x,y)
```

```python
#plotting only x
plt.plot(x)
```

```python
%matplotlib inline

#for figure size
plt.figure(figsize=(10,10))

#bar plot
plt.bar(x,y,color='b')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('bar graph')
```

```python
x1 = [10,20,30,40,50,60]
y1 = [23,11,9,37,12,10]
x2 = [15,25,35,45,55,65]
y2 = [22,55,6,7,99,7]
```

```python
#'go' is used for green c ircles and 'r^' is used for red triangles
plt.plot(x1,y1,'go',x2,y2,'r^')
```

```python
plt.bar(x1,y1,width=4)
plt.bar(x2,y2,width=4)
plt.legend()
plt.show()
```

```python
#The subplot() method takes three arguments: number of rows, number of columns and index
plt.subplot(1,2,1)
plt.plot(x1,y1,'go')

plt.subplot(1,2,2)
plt.plot(x2,y2,'r^')

plt.show()
```

```python
#The subplot() method takes three arguments: number of rows, number of columns and index
plt.subplot(2,1,1)
plt.plot(x1,y1,'go')

plt.subplot(2,1,2)
plt.plot(x2,y2,'r^')

plt.show()
```

```python
#scatter plot
plt.scatter(x,y,color='r')
```
